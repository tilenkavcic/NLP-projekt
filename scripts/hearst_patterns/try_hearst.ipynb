```python
TP = 0
TN = 0
FP = 0
FP = 0

for index, row in df.iterrows():
    #print(row['DEFINIENDUM'], row['GENUS'], row['SENTENCE'])
    #print(h.hearstify(row['SENTENCE']))
    if type(row['GENUS']) == float or len(row['GENUS']) <= 0:
        continue
    genuses = row['GENUS'].lower().replace(" ", "").split("|")
    result = h.hearstify(row['SENTENCE'])
    for i in result:
        #print(i[0].lower(), "=", i[1].lower().replace(" ", ""), ".....",
        #row['DEFINIENDUM'].lower(), "=",genuses)
        if i[0].lower() == row['DEFINIENDUM'].lower():
            if i[1].lower().replace(" ", "") in genuses:
                TP += 1
            else:
                FP += 1
        else:
            FP += 1
print("tp:",TP,"\nFP:",FP)
```
